```python
data = pd.read_excel('https://archive.ics.uci.edu/ml/machine-learning-databases/00350/default%20of%20credit%20card%20clients.xls', index_col=0)
```

```python
data.info()
```

```python
data.columns = data.iloc[0]
```

```python
data.drop(data.index[0], inplace=True)
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.describe()
```

```python
data = data.apply(pd.to_numeric)
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
mask = np.random.random(data.shape) < 0.25
```

```python
data_with_nans = data.mask(mask)
```

```python
data_with_nans.isna().sum()/len(data)
```

```python
data_with_nans.head()
```

```python
non_nans = data_with_nans.dropna()
```

```python
non_nans
```

```python
print(f"The df had {len(data_with_nans)} rows, but after we dropped all rows containing NaNs, \
it was reducted to {len(non_nans)} rows.")
```

```python
data_with_nans.mean()
```

```python
filled_with_mean = data_with_nans.fillna(data_with_nans.mean())
```
